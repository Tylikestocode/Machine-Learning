# Breast Cancer Classification Using Logistic Regression

```python
import pandas as pd
```

# Overview of the dataset

```python
data = pd.read_csv('breast-cancer.csv')
data.head()
```

# Data Preprocessing

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python
Train, Test = train_test_split(data, random_state=42, test_size=0.2)
```

```python
X_train = Train.drop(['id', 'diagnosis'], axis=1)
y_train = Train['diagnosis']

print(f'X Train Shape : {X_train.shape}')
print(f'y Train Shape : {y_train.shape}')
```

```python
X_test = Test.drop(['id', 'diagnosis'], axis=1)
y_test = Test['diagnosis']

print(f'X Test Shape : {X_test.shape}')
print(f'y Test Shape : {y_test.shape}')
```

```python
class DataPipeLine():
    
    def __init__(self, data):
        self.data = data
        self.num_pipeline = make_pipeline(StandardScaler())

    def transform(self):

        num_features = self.data.select_dtypes(['int', 'float']).columns
        pipeline = ColumnTransformer([
            ('num', self.num_pipeline, num_features)
        ])

        return pd.DataFrame(pipeline.fit_transform(self.data), columns=pipeline.get_feature_names_out())
        
```

```python
X_train = DataPipeLine(X_train).transform()
X_test = DataPipeLine(X_test).transform()
```

```python
target_map = {
    'M' : 1,
    'B' : 0
}

y_train = y_train.map(target_map)
y_test = y_test.map(target_map)
```

# Model Selection

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
```

```python
LGR = LogisticRegression()

LGR_params = {
    'solver' : ['saga', 'liblinear'],
    'penalty' : ['l1', 'l2'],
    'max_iter' : [10000]
}

LGR_grid = GridSearchCV(LGR, param_grid=LGR_params, cv=5)
LGR_grid.fit(X_train, y_train)
```

```python
print(f'Best Score: {LGR_grid.best_score_}')
print(f'Best Params: {LGR_grid.best_params_}')
```

```python
LGR = LogisticRegression(penalty='l1', solver='saga', max_iter=10000)
LGR.fit(X_train, y_train)
```

# Model Evaluation

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
test_pred = LGR.predict(X_test)
```

```python
f1 = f1_score(y_true=y_test, y_pred=test_pred)
prec_score = precision_score(y_true=y_test, y_pred=test_pred)
acc = accuracy_score(y_true=y_test, y_pred=test_pred)
recall = recall_score(y_true=y_test, y_pred=test_pred)

print('Classfication Report:')
print('---------------------')
print(f'F1 : {f1:.2f}')
print(f'Precision : {prec_score:.2f}')
print(f'Recall : {recall:.2f}')
print(f'Accuracy : {acc:.2f}')
print('---------------------')
```

```python
conf_matrix = confusion_matrix(y_true=y_test, y_pred=test_pred)
sns.heatmap(conf_matrix, annot=True)
plt.title('Confusion Matrix Results on Test Set')
plt.show()
```
